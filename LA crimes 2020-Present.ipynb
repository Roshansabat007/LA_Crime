```python
#Filling up the missing values

df['Mocodes'].fillna('Unknown', inplace=True)
df['Mocodes'].isnull().sum()
```

```python
df['Vict Sex'].fillna('Unknown', inplace=True)
```

```python
df['Vict Descent'].fillna('Unknown', inplace=True)
```

```python
df['Premis Cd'].fillna(df['Premis Cd'].mode()[0], inplace=True)
```

```python
df['Premis Desc'].fillna('Unknown', inplace=True)
```

```python
df['Weapon Used Cd'].fillna(0, inplace=True)
```

```python
df['Weapon Desc'].fillna('No weapon', inplace=True)
```

```python
df['Status'].fillna(df['Status'].mode()[0], inplace=True)
```

```python
df['Premis Cd'].fillna(df['Premis Cd'].mode()[0], inplace=True)
```

```python
df['Crm Cd 1'].fillna(df['Crm Cd 1'].mode()[0], inplace=True)
df['Crm Cd 2'].fillna(0, inplace=True)
df['Crm Cd 3'].fillna(0, inplace=True)
df['Crm Cd 4'].fillna(0, inplace=True)
```

```python
df['Cross Street'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'].str.split().str[0], format='%m/%d/%Y', errors='coerce')
```

```python
df['Date Rptd'].head()
```

```python
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'].str.split().str[0], format='%m/%d/%Y', errors='coerce')
```

```python
df['DATE OCC'].head()
```

```python
df['TIME OCC'] = df['TIME OCC'].astype(str).str.zfill(4)
hours = df['TIME OCC'].str[:2].astype(int)
minutes = df['TIME OCC'].str[2:].astype(int)

df['DATETIME OCC'] = df['DATE OCC'] + pd.to_timedelta(hours, unit="h") + pd.to_timedelta(minutes, unit="m")
```

```python
df['DATETIME OCC'].head()
```

```python
print(df.duplicated().sum())
```

```python
df['Vict Age'].unique()
```

```python
df = df[(df["Vict Age"] > 0) & (df["Vict Age"] < 120)]
```

```python
df['Vict Age'].unique()
```

```python
df["Hour"] = df["DATETIME OCC"].dt.hour
df["Day of Week"] = df["DATETIME OCC"].dt.day_name()
df["Month"] = df["DATETIME OCC"].dt.month_name()
```

VISUALIZATION

```python
ncoa=df['AREA NAME'].value_counts()
ncoa
```

```python
plt.figure(figsize=(16,8))
plt.pie(ncoa.values, labels=ncoa.index, autopct='%1.1f%%', colors=['pink','red','blue','purple','gold'])
plt.show
```

```python
df["Year-Month"] = df["DATETIME OCC"].dt.to_period("M").astype(str)
plt.figure(figsize=(12,6))
sns.lineplot(data=df["Year-Month"].value_counts().sort_index(), marker="o")

plt.title("Crime Trend Over Time")
plt.xlabel("Month-Year")
plt.ylabel("Number of Crimes")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df["Hour"].value_counts().index, y=df["Hour"].value_counts().values, palette="coolwarm")

plt.title("Crimes by Hour of the Day")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Crimes")
plt.xticks(range(0,24))
plt.grid(axis="y")
plt.show()
```

```python
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
plt.figure(figsize=(12,6))
sns.barplot(x=df["Day of Week"].value_counts().reindex(day_order).index,
            y=df["Day of Week"].value_counts().reindex(day_order).values, palette="viridis")

plt.title("Crimes by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Number of Crimes")
plt.grid(axis="y")
plt.show()
```

```python
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", "December"]

plt.figure(figsize=(12,6))
sns.barplot(x=df["Month"].value_counts().reindex(month_order).index,
            y=df["Month"].value_counts().reindex(month_order).values, palette="magma")

plt.title("Crimes by Month")
plt.xlabel("Month")
plt.ylabel("Number of Crimes")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(y=df["Crm Cd Desc"].value_counts().nlargest(10).index, 
            x=df["Crm Cd Desc"].value_counts().nlargest(10).values, palette="Blues_r")

plt.title("Top 10 Most Common Crimes")
plt.xlabel("Number of Crimes")
plt.ylabel("Crime Type")
plt.grid(axis="x")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df["Vict Age"], bins=30, kde=True, color="green")

plt.title("Distribution of Victim Ages")
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data= df, x="Vict Age")

plt.title("Distribution of Victim Ages")
plt.xlabel("Age")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(y=df["Premis Desc"].value_counts().nlargest(10).index, 
            x=df["Premis Desc"].value_counts().nlargest(10).values, palette="rocket")

plt.title("Top 10 Locations of Crimes")
plt.xlabel("Number of Crimes")
plt.ylabel("Premises Type")
plt.grid(axis="x")
plt.show()
```

```python
weapon_counts = df["Weapon Desc"].value_counts().nlargest(5)
plt.figure(figsize=(8,8))
plt.pie(weapon_counts, labels=weapon_counts.index, autopct="%1.1f%%", colors=sns.color_palette("pastel"))

plt.title("Crimes Involving Weapons")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x=df["Part 1-2"], palette="coolwarm")

plt.title("Distribution of Part 1 and Part 2 Crimes")
plt.xlabel("Crime Category")
plt.ylabel("Number of Crimes")
plt.xticks(ticks=[0,1], labels=["Part 1 (Serious Crimes)", "Part 2 (Less Serious Crimes)"])
plt.grid(axis="y")

plt.show()
```
